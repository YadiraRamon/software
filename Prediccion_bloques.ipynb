```python
#1_Dataset
df = pd.read_csv('Composito.csv', sep=';')
df.head()
```

```python
#Variables dependientes
# Especifica los nombres de las columnas
columnas_de_interes = ['Este', 'Norte', 'Elevacion']  # Ajusta las columnas segn tu DataFrame
# Crea un nuevo DataFrame con las columnas seleccionadas
Features_pred = df[columnas_de_interes]
print(Features_pred.shape)
```

```python
# Normalizacion
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
normalized_Features_pred = scaler.fit_transform(Features_pred)
normalized_Features_pred
```

```python
#Selecciona las variables independientes (features)
X = df[columnas_de_interes] # Agrega aqu todas tus variables independientes

# Crea un conjunto de datos sinttico para Y (puedes ajustar esto segn la relacin que desees)
Y = np.sin(X['Este']) + np.random.normal(0, 0.1, len(X))

# Crea y entrena el modelo SVR
svr_model = SVR(kernel='linear')  # Puedes ajustar el kernel segn tus necesidades
svr_model.fit(X, Y)

# Realiza predicciones
predictions = svr_model.predict(X)
predictions
```
